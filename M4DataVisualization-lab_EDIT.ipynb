## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# inpect the database

QUERY = """
SELECT *
FROM sqlite_master
"""

df = pd.read_sql_query(QUERY,conn)
df
```

```python
QUERY = """
SELECT *, count(*) as Total
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY Total Desc
LIMIT 5
"""
def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n({:d} respondent)".format(pct, absolute)

df = pd.read_sql_query(QUERY,conn)
plt.figure(figsize=(10,9))

plt.pie(x=df['Total'], labels=df['DatabaseDesireNextYear'], autopct=lambda pct: func(pct, df['Total']), explode=(0.08, 0.01, 0.01, 0.01, 0.01))
plt.legend()
plt.show()
```

In the list of most popular languages respondents wish to learn next year, what is the rank of Python?

```python
QUERY = """
SELECT *, count(*) as Total
FROM LanguageDesireNextYear
GROUP BY LanguageDesireNextYear
ORDER BY Total Desc
LIMIT 5
"""
def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n({:d} respondent)".format(pct, absolute)

df = pd.read_sql_query(QUERY,conn)
plt.figure(figsize=(10,9))

plt.pie(x=df['Total'], labels=df['LanguageDesireNextYear'], autopct=lambda pct: func(pct, df['Total']), explode=(0.08, 0.01, 0.01, 0.01, 0.01))
plt.legend()
plt.show()
```

How many respondents indicated that they currently work with 'SQL'?

```python
QUERY = """
SELECT *, count(*) as Total
FROM LanguageWorkedWith
GROUP BY LanguageWorkedWith
ORDER BY Total Desc
LIMIT 5
"""
def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n({:d} respondent)".format(pct, absolute)

df = pd.read_sql_query(QUERY,conn)
plt.figure(figsize=(12,9))

plt.pie(x=df['Total'], labels=df['LanguageWorkedWith'], autopct=lambda pct: func(pct, df['Total']), explode=(0.08, 0.01, 0.01, 0.01, 0.01))
plt.legend()
plt.show()
```

How many respondents indicated that they work on 'one language' and 'MySQL' only? 

```python
QUERY = """
SELECT * 
FROM DatabaseWorkedWith
ORDER BY Respondent
"""

df = pd.read_sql_query(QUERY,conn)

#For respondent with one language only
one_only_counts = df.groupby('Respondent')['DatabaseWorkedWith'].nunique()
one_only_counts = one_only_counts[one_only_counts == 1]

#For respondent with MySQL only
mysql_only = df.groupby('Respondent')['DatabaseWorkedWith'].unique()
mysql_only = pd.DataFrame(mysql_only).reset_index()

#1st method, for faster lambda can be used (remove 2nd option until #Counting section)
#mysql_only_count = mysql_only[mysql_only['DatabaseWorkedWith'].apply(lambda x: len(x) == 1 and x[0] == 'MySQL')]

#2nd method, for conventional iteration method use .iterrows()
mysql_only_count = []
for idx, row in mysql_only.iterrows():
    if len(row['DatabaseWorkedWith']) == 1 and row['DatabaseWorkedWith'][0] == 'MySQL':
        mysql_only_count.append(row)

#COUNTING
mysql_only = len(mysql_only_count)

print('There are ', len(one_only_counts), ' respondents who worked with one language only')
print('There are ', mysql_only, ' respondents who worked with MySQL only')
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
WHERE Age between 30 and 35
"""

df = pd.read_sql_query(QUERY,conn)
df
plt.figure(figsize=(7,6))
wwh = df['WorkWeekHrs'].quantile(0.5)
crh = df['CodeRevHrs'].quantile(0.5)

plt.bar('Age Group 30 to 35', wwh, label='WorkWeekHrs')
plt.bar('Age Group 30 to 35', crh, label='CodeRevHrs')
plt.title('Stacked Bar of WorkWeekHrs and CodeRevHrs Median for Age Group 30 to 35')
plt.legend()
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = """
SELECT ConvertedComp, Age
FROM master
WHERE Age between 45 and 60
"""

df = pd.read_sql_query(QUERY,conn)
plt.figure(figsize=(11,8))

ccmed = df.groupby('Age')['ConvertedComp'].median()
ccmed = pd.DataFrame(ccmed).reset_index()

plt.plot(ccmed['Age'], ccmed['ConvertedComp'], linestyle='-', marker='o', label='Median ConvertedComp')
plt.title('Line Chart of ConvertedComp Median for Age Group 45 to 60')
plt.xlabel('Age Group')
plt.ylabel('Median ConvertedComp')
plt.grid(True) #for better readability
plt.legend()
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
from matplotlib.patches import Patch

QUERY = """
SELECT MainBranch, count(*) as count
FROM master
GROUP BY MainBranch
"""

df = pd.read_sql_query(QUERY,conn)
plt.figure(figsize=(11,8))
colors = ['orange','cyan']
plt.bar(df['MainBranch'], df['count'], color=colors)

plt.title('Bar Chart of Main Branch Distribution')
patches = [Patch(color=color) for color in colors]
plt.legend(patches, df['MainBranch'].tolist(), title='Main Branch Distribution')
plt.show()
```
